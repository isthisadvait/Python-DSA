```python
#Implementation of HashMap using inbuilt hash function
class MapNode:
    def __init__(self,key,value) :
           self.key=key
           self.value=value
           self.next=None
    
class Map:
    def __init__(self) :
        self.Bsize=10
        self.B=[None for i in range(self.Bsize)]
        self.count=0
   
    def size(self):
        return self.count
    def LoadFactor(self):
        return (self.count/self.Bsize)

    def rehash(self):
        temp=self.B
        self.B=[None for i in range(2*(self.Bsize))]
        self.Bsize=2*(self.Bsize)
        self.count=0
        for head in temp:
            while head is not None:
                self.insert(head.key,head.value)
                head=head.next

    
    def insert(self,key,value):
        hc=hash(key) #hash function give hash code
        idx=abs(hc)%(self.Bsize) #Compressing to fit in required range
        head=self.B[idx]
        while head is not None:
            if head.key==key:
                head.value=value
                return
            else :
                head=head.next
        head=self.B[idx]
        newNode = MapNode(key,value)
        newNode.next=head
        self.B[idx]=newNode
        self.count+=1
        loadFactor=self.count/self.Bsize
        if loadFactor>=0.7:
            self.rehash()

    def search(self,key):
        hc=hash(key) #hash function give hash code
        idx=abs(hc)%(self.Bsize) #Compressing to fit in required range
        head=self.B[idx]
        while head is not None:
            if head.key==key:
                return head.value
            else :
                head=head.next
        return None

    def remove(self,key):
        hc=hash(key) #hash function give hash code
        idx=abs(hc)%(self.Bsize) #Compressing to fit in required range
        head=self.B[idx]
        prev=None
        while head is not None:
            if head.key==key:
                if prev is None:
                    self.B[idx]=head.next
                    head=None
                    self.count-=1
                    del head
                else :
                    prev.next=head.next
                    self.count-=1
                    head=None
                    del head
            else:
                prev=head
                head=head.next
        return


        
```
